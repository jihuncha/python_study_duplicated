```python
# 플로이드 워셜

INF = int(1e9)

# 노드및 간선
n = int(input())
m = int(input())

# 2차원리스트 -> 처음에 INF 로 초기화
graph = [[INF] * (n + 1) for _ in range(n + 1)]

# print(graph)

for a in range(n+1):
    for b in range(n+1):
        if a==b:
            graph[a][b] = 0
# print(graph)

for _ in range(m):
    a,b,c = map(int,input().split())
    graph[a][b] = c
    
for k in range(1, n+1):
    for a in range(1, n+1):
        for b in range(1, n+1):
            graph[a][b] = min(graph[a][b], graph[a][k] + graph[k][b])
            
for a in range(1, n+1):
    for b in range(1, n+1):
        if graph[a][b] == INF:
            print("no!! INF!!", end=' ')
        else:
            print(graph[a][b], end=' ')
    print()
    
```
